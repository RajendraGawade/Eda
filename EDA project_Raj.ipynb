# Missing value Treatment

```python
data['PerformanceRating']=data['PerformanceRating'].fillna(data['PerformanceRating'].mean())
```

```python
data['DistanceFromHome']=data['DistanceFromHome'].fillna(data['DistanceFromHome'].mean())
```

```python
data['JobLevel'].bfill(axis=0,inplace=True)
```

```python
data['MonthlyIncome']=data['MonthlyIncome'].fillna(data['MonthlyIncome'].mean())
```

```python
data['NumCompaniesWorked'].ffill(axis=0,inplace=True)
```

```python
data['OverTime'] = data['OverTime'].fillna(data['OverTime'].mode(0).iloc[0])
```

```python
data['PercentSalaryHike']=data['PercentSalaryHike'].fillna(data['PercentSalaryHike'].mean())
```

```python
data['TotalWorkingYears'].ffill(axis=0,inplace=True)
```

```python
data['TrainingTimesLastYear']=data['TrainingTimesLastYear'].fillna(data['TrainingTimesLastYear'].mode().iloc[0])
```

```python
data['WorkLifeBalance'] = data['WorkLifeBalance'].replace(np.nan, np.min)
```

```python
data['YearsAtCompany'].ffill(axis=0,inplace=True)
```

```python
data['YearsSinceLastPromotion'] = data['YearsSinceLastPromotion'].replace(np.nan,np.min)
```
